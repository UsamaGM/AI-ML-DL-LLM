```python
import matplotlib.pyplot as plt

year = [1950, 1970, 1990, 2020]
pop = [2.245, 1.223, 2.455, 5.552]

plt.plot(year, pop)
plt.show()
```

```python
plt.scatter(year, pop)
plt.show()
```

```python
values = [0, 0.6, 0.9, 1, 2.4, 2.2, 2, 2.5, 2.9, 5, 0.5, 5.5, 8, 6.5, 4.4, 4.8]
plt.hist(values, bins=2)
plt.show()
```

### When to use which
- Use line plot when seeing rate of change of one with respect to another.
- Use Scatter plot when seeing if there is an underlying relation between two quantities.
- Use Histogram to see the distribution of data.
